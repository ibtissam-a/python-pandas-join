```python
import pandas as pd 
```

```python
# first we will create our data frames 
df1 = pd.DataFrame({'Name':['Ali','Mohamed','Sonia','Alex','Vedet','Rakish'],
                    'From':['Tunisia','Syria','Egypt','Canada','Turkey','India'],
                    'Age':['26','30','19','40','10','35'],
                    'Languages':['Arabic','Arabic','Arabic','English','Turc','Hindi']
                    })

df2 = pd.DataFrame({'Name':['Ali','Omar','Sonia','Rakish'],
                    'High':['190','186','175','180'],
                    'Weight':['85','90','70','83']
                    })

df3 = pd.DataFrame({'Persons':['Ali','Mostapha','Sabrine','Hala','Vedet','Rakish'],
                    'Grade':['Data_scientist','Police','Doctor','Teacher','Acter','Data_nalyst'],
                    'Gendre':['Male','Male','Female','Female','Male','Male']
                    })
```

```python
# Inoreder to create all joins, we will use the pandas function "merge" with the folowing syntax:
# pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=True)
```

# Inner join(intersection)
* Atleast one common column has the same values in the two data frames
* The resultant df will have: 
    * All columns of the merged dfs
    * Only rows with common values 
    


```python
inner = pd.merge(df1, df2,how = 'inner',on = 'Name')
inner
```

# Outer Join
* Atleast one common column has the same values in the two data frames
* The resultant df will have: 
    * All columns of the merged dfs
    * All rows , empty cells will be fill with NaN 


```python
outer = pd.merge(df1,df2, how = 'outer', on = 'Name', sort= True)
```

```python
outer
```

# Left join
* Columns name are different but there are some common values
* The resultant df will have: 
    * All columns of the merged dfs
    * All rows from the left df + common rows from the right df 


```python
left = pd.merge(df1,df3,how='left', left_on = 'Name', right_on='Persons', indicator=True)
left
```

```python
# drope all columns with NaN values
left.dropna(axis=1)
```

```python
# drope all rows with NaN values
left.dropna(axis=0)
```

# Right join
* Atleast one common column with some common values
* The resultant df will have: 
    * All columns of the merged dfs
    * All rows from the right df + common rows from the left df 


```python
# the new data frame will have the same number of rows of the data frame on the right 
# coz we will keep the dataframe on the right and we will add only the columns of the left data frame with common values 
right = pd.merge(df1, df2, how ='right', on = 'Name', indicator=True)
right
```

```python
# the real shape of the resultant data frame is: (nb_rows = nb_rows_right_df,  nb_col = nb_col_right_df + nb_col_left_df-1) 
# (the 1 is for the common column)
right.shape
```
